## 크롤링

````python
import requests
from bs4 import BeautifulSoup

def crawl_article(url):
    try:
        res = requests.get(url, timeout=5, headers={"User-Agent": "Mozilla/5.0"})
        res.raise_for_status()
        soup = BeautifulSoup(res.text, "lxml")

        # 네이버 뉴스 기본 본문
        article = soup.find("div", id="dic_area")
        if article:
            return article.get_text(strip=True, separator="\n")

        # fallback (다른 언론사)
        paragraphs = soup.find_all("p")
        if paragraphs:
            return "\n".join(p.get_text(strip=True) for p in paragraphs)

        return None
    except Exception as e:
        print(f" 크롤링 실패: {url} ({e})")
        return None


#  위 셀에서 얻은 data 사용
for idx, item in enumerate(data['items'], 1):
    url = item['link']
    text = crawl_article(url)

    print(f"[{idx}] {item['title']}")
    print(url)
    if text:
        print("본문 길이:", len(text), "자")
        print(text, "\n")   # 전체 본문 출력
    else:
        print(" 본문 없음\n")
````

````python
articles = []

for idx, item in enumerate(data['items'], 1):
    url = item['link']
    text = crawl_article(url)   # 본문 텍스트 수집
    if text:
        articles.append({
            "title": item['title'],
            "url": url,
            "content": text
        })
````

## 본문에 대한 QA pair 만들기

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

news_qa_prompt = PromptTemplate.from_template("""
                                              
당신은 뉴스 기사 내용을 바탕으로 퀴즈를 출제하는 교수입니다. 
아래 컨텍스트(뉴스 기사 본문)만을 기반으로 퀴즈 문항과 정답을 생성하세요.
외부 지식은 절대 사용하지 마세요. 

---------------------
{context}
---------------------

규칙:
1. 반드시 {num_questions}개의 퀴즈를 만들 것
2. 모든 질문과 답변은 한국어로 작성
3. 답변에는 반드시 위 기사 본문에서 문장을 일부 그대로 인용하여 출처를 명시할 것
4. 출력은 JSON 객체들의 나열로만 할 것 (배열 [] 사용 금지)
5. 답변은 원문 의미를 충실히 반영하되, 그대로 복붙하지 말고 비전공자의 눈높이에 맞춰 간결하거나 자연스럽게 바꿔라.  
6. 필요할 경우 간단한 배경 설명을 붙여 문해력을 돕는다.                                           

# JSON 출력 예시:
```json
{{
  "QUESTION": "삼성전자가 개최한 포럼의 명칭은 무엇입니까?",
  "ANSWER": "삼성전자가 개최한 포럼은 제9회 삼성 AI 포럼 2025입니다.",
  "SOURCE": "삼성전자는 제9회 삼성 AI 포럼 2025를 15일부터 개최한다."
}},
{{
  "QUESTION": "첫날 포럼에서 논의된 주제는 무엇입니까?",
  "ANSWER": "첫날 포럼에서는 반도체 산업의 버티컬 AI 전략이 논의되었습니다.",
  "SOURCE": "첫날 포럼에서는 반도체 산업의 버티컬 AI 전략이 논의되었다."
}}

"""
)
````

````python
from langchain_openai import ChatOpenAI

# LLM 초기화
llm = ChatOpenAI(model="gpt-5")

# 체인 구성
chain = news_qa_prompt | llm | parser

# 모든 기사에 대해 반복 실행
results = []
for idx, article in enumerate(articles, 1):
    result = chain.invoke({
        "context": article["content"],  # 각 기사 본문 전체
        "num_questions": 2              # 기사마다 2문제 생성
    })
    results.append({
        "index": idx,
        "title": article["title"],
        "url": article["url"],
        "qa": result
    })

# 결과 출력
for r in results:
    print(f"[{r['index']}] {r['title']}")
    print(f"URL: {r['url']}")
    print(r["qa"])
    print("="*60)
````

## jsonl 저장

````python
import json
import re

with open("qa_output.jsonl", "w", encoding="utf-8") as f:
    for r in results:
        qa_raw = r["qa"]
        cleaned_qas = []

        if isinstance(qa_raw, str):
            # 중괄호 블록만 추출
            json_objects = re.findall(r"\{.*?\}", qa_raw, flags=re.DOTALL)
        elif isinstance(qa_raw, list):
            json_objects = qa_raw
        else:
            json_objects = []

        for obj in json_objects:
            try:
                qa = json.loads(obj) if isinstance(obj, str) else obj

                if not isinstance(qa, dict):
                    continue

                # SOURCE 제거
                cleaned = {
                    "QUESTION": qa.get("QUESTION", "").strip(),
                    "ANSWER": qa.get("ANSWER", "").strip()
                }

                # 저장 (한 줄에 하나)
                f.write(json.dumps(cleaned, ensure_ascii=False) + "\n")

            except Exception as e:
                print(f"⚠️ 파싱 실패: {e} → {obj}")
                continue

print("jsonl 저장 완료: qa_output.jsonl")
````
